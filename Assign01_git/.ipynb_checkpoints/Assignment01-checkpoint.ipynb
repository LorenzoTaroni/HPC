```python
import numpy as np
import matplotlib.pyplot as plt
import csv
plt.style.use("ggplot")
```

## Section  1: theoretical model

```python
T_comp = 2*10**(-9)
T_read = 10**(-4)
T_comm = 10**(-6)
P = np.array(range(1,1001))
#rcParams["figure.figsize"] = [6.4, 4.8].
plt.rcParams["figure.figsize"] = [10, 5]
```

```python
plt.xlabel('Number of processors P')
plt.ylabel('SpeedUp(P)')

for i in range(6,12):
    
    n = 10**(i)

    T1 = T_comp*n + T_read

    TP = T_comp*(P-1+n/P) + T_read + 2*(P-1)*T_comm
    
    #print(P[(T1/TP).argmax()])
          
    _= plt.plot(P, T1/TP)
    
    _= plt.axvline(x = P[(T1/TP).argmax()], ymin = 0, ymax = max(T1/TP))
```

## Section 2 : play with MPI program

```python
def load_data(file,n):
    """Returns a numpy array"""
    data = []
    with open(file) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        for line in csv_reader:
            data += [float(line[n])]
       
    return(np.asarray(data[1:]))
```

```python
def max_time(file,n):
        data = []
        with open(file) as csv_file:
            csv_reader = csv.reader(csv_file, delimiter=',')
            for line in csv_reader:
                data += [float(line[n])]
        return(max(np.asarray(data)))
```

### 2.1:  compute strong scalability of a mpi_pi.c program

```python
x = load_data('time_elaps_strong.txt',0)
ys_elaps = load_data('time_elaps_strong.txt',1)
```

```python
plt.xlabel('Number of Processors P')
plt.ylabel('Elapsed Time')
_ = plt.scatter(x,ys_elaps)
_ = plt.plot(x,ys_elaps)
```

```python
time = []
time += [max_time('walltime_1_strong.txt',1)]
time += [max_time('walltime_2_strong.txt',1)]
time += [max_time('walltime_4_strong.txt',1)]
time += [max_time('walltime_8_strong.txt',1)]
time += [max_time('walltime_16_strong.txt',1)]
time += [max_time('walltime_20_strong.txt',1)]
ys_walltime = np.asarray(time)
```

```python
plt.xlabel('Number of Processors P')
plt.ylabel(' Walltime')
_ = plt.scatter(x,ys_walltime)
_ = plt.plot(x,ys_walltime)
```
